## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10))
with sns.axes_style("whitegrid"):

  grafico= sns.boxplot(x=data['price'])
  grafico.set(xlabel= "Preços")
  grafico.set_title("Distribuição dos diamantes por preços", fontsize=15, fontweight="bold")
  grafico.set_xlabel("Preço (USD)", fontsize=15);
  
```

```
# Insight do gráfico 1: 
# A maioria dos preços dos diamantes estão concentrados entre 500 e 5300 aproximadamente! Mas se vê que há diamantes mais baratos e mais caros também. A mediana está em quase 2500!
# O valor dos diamantes se estendem até 12000 aproximadamente e dificilmente serão encontrados diamantes mais caros em grandes quantidades!
```

```
# gráfico do exercício 1
import matplotlib.pyplot as plt
plt.figure(figsize=(23,10))
with sns.axes_style("whitegrid"):

  grafico= sns.boxplot(x=data['price'], y=data['cut'] )
  grafico.set_title("Distribuição dos diamantes por preço separados por corte", fontsize=15, fontweight="bold")
  grafico.set_xlabel("Preço (USD)", fontsize=15);
  grafico.set_ylabel("Corte", fontsize=15);
  
  # Outro Insight do gráfico 1: 
  # Aqui fiz a concentração de diamantes separando por corte. 
  # Dá pra visualizar por exemplo o diamante com o tipo de corte Fair, que possui valores de maior concentração entre 2000 e 5200 aproximadamente!
  # O valor pode se estender até 10000 e dificilmente será encontrado um diamante com o corte Fair em grandes quantidades com valor superior a 10000!
  # O mesmo raciocínio serve para analisar os preços dos outros tipos de cortes também!
```

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
diamonds = data[['carat','price','clarity']].groupby("clarity")
diamonds.head()
```

```
# gráfico do exercício 2

import matplotlib.pyplot as plt
plt.figure(figsize=(25,21))

with sns.axes_style("whitegrid"):
  grafico = sns.scatterplot(data=data, x= 'price', y='carat', hue= 'clarity',style='clarity', palette="pastel")
  grafico.get_legend().set_title('claridade');
  grafico.set_title("Preço dos diamantes por seu peso", fontsize=15, fontweight="bold")
  grafico.set_xlabel("Preço(USD)", fontsize=15);
  grafico.set_ylabel("Peso (Quilate)", fontsize=15);
```

**Insight do gráfico 2**: ...

```
#Insight do gráfico 2:

# Aqui podemos observar que a maioria dos diamantes por claridade se concentram entre 1 e 2 quilates com preços que vão até mais de 18000 e muita variedade nas claridades!
# Até 1 Quilate os preços vão até 10000 dependendo da claridade.
# Entre 1 e 2 quilates temos diamantes entre 1250 a 18000 aproximadamente dependendo da claridade.
# Entre 2 e 3 quilates vemos muita concentração de diamantes com claridade VS2 e SL2 com preços que vão de 5000 a 18000 aproximadamente.
# Também se percebe que quanto maior é o quilate do diamante sua quantidade diminui 
# Os diamantes de 3 e 5 quilates são os mais caros e estão nas claridades I1, SI2, SI1 e algum VS2.
# Há poquíssimas unidades de diamantes entre 4 e 5 quilates.
```

```
# Aqui eu filtrei por claridade IF! Podemos ter uma visão melhor dos preços por claridade em relação ao peso. 
import matplotlib.pyplot as plt
diamonds = data[data['clarity'] == 'IF'] #Filtro
plt.figure(figsize=(20,14))

with sns.axes_style("whitegrid"):
  grafico = sns.scatterplot(data=diamonds, x= 'price', y='carat', hue= 'clarity',style='clarity', palette="pastel")
  grafico.get_legend().set_title('Claridade');
  grafico.set_title("Preço dos diamantes por seu peso separado pela claridade IF", fontsize=15, fontweight="bold")
  grafico.set_xlabel("Preço(USD)", fontsize=15);
  grafico.set_ylabel("Peso (Quilate)", fontsize=15);

# Outro Insight do gráfico 2: 
# Até 0.5 quilates por exemplo o preço vai até 2600 aproximadamente
# 0.5 até 1.0 varia entre 1000 e 9000 aproximadamente
# 1.0  até 1.5 varia entre 5200 e 18000 aproximadamente
# 1.5 até 2.0 varia entre 8000 e 18000 aproximadamente
# 2.0 em diante varia entre 16.800 e 18000 aproximadamente
```

```
# Outro Insight do gráfico 2: 
# Usando o gráfico Histplot para uma análise em relação a quantidade de diamantes pela claridade baseado no corte!

import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))

diamonds = data[['carat','price','cut','clarity']]
with sns.axes_style('whitegrid'):
  
  grafico = sns.histplot(data=diamonds, x="clarity",  hue="cut",kde=True);
  grafico.set_title("Quantidade dos diamantes pela claridade baseados no corte!",fontsize=15, fontweight="bold");
  grafico.set_xlabel("Claridade", fontsize=15);
  grafico.set_ylabel("Contagem", fontsize=15);
  grafico.get_legend().set_title('corte');
```

```
# Usando o Facetgrid para uma análise em relação a quantidade de diamantes por corte e clareza!
# Separando por colunas temos uma visão melhor para fazermos a análise!
#Falta por o título.
import matplotlib.pyplot as plt
plt.figure(figsize=(25,15))

diamonds = data[['carat','price','cut','clarity']]
with sns.axes_style('whitegrid'):

  
  grafico = sns.FacetGrid(data=diamonds, col="cut",  hue="cut")
  grafico.map(sns.histplot, "clarity", kde=True)
  grafico.fig.suptitle('Quantidade dos diamantes por corte e claridade',fontsize=15, fontweight="bold")
  grafico.fig.subplots_adjust(top= 0.8)

  altura  = 5
  largura = 19

  grafico.fig.set_size_inches(w=largura, h=altura)

  

# Outro Insight do gráfico 2: 
# Aqui observamos que a quantidade maior de diamantes por corte é o ideal e com clareza VS2! 
# Percebemos também a diminuição da quantidade de diamantes em relação ao seu corte!
# O corte Premium vem em segundo lugar em quantidade mas pode perder por exemplo na claridade WS1, para os diamantes com o corte Very Good!
# Achei legal essa análise também! Ficou bem visual.
```

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10))
with sns.axes_style("whitegrid"):

  cars = sns.boxplot(x=data['weight'])
  
  cars.set_title("Distribuição contínua aproximada do peso de carros", fontsize=15, fontweight="bold")
  cars.set_xlabel("Peso", fontsize=15);

# Insight do gráfico 3
# Aqui Observamos que a maior concentração de carros são aqueles com  peso entre  2100 e 3600, estendendo-se até 5300 aproximadamente.
```

```
# gráfico do exercício 3
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10))
with sns.axes_style("whitegrid"):

  cars = sns.boxplot(x=data['weight'],y=data['origin'] )
  
  cars.set_title("Distribuição contínua aproximada do peso de carros por País", fontsize=15, fontweight="bold")
  cars.set_xlabel("Peso", fontsize=15);
  cars.set_ylabel("Origem(País)", fontsize=15);
  
  # Insight do gráfico 3
  # Grande parte dos carros mais pesados são produzidos nos EUA
  # Os carros mais leves no japão
  # Os carros com pesos mais leves e um pouco mais pesados na Europa.
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10))
with sns.axes_style("whitegrid"):

  cars = sns.boxplot(x=data['cylinders'],y=data['weight'] )
  
  cars.set_title("Distribuição contínua aproximada do peso de carros por Cilindros", fontsize=15, fontweight="bold")
  cars.set_xlabel("Cilindros", fontsize=15);
  cars.set_ylabel("Peso", fontsize=15);

  # Insight do gráfico 3
  #Quanto mais cilindros maior é o peso do carro podendo haver carros com 4 cilindros mais leves do que os de 3! Mas a tendência é que pesem mais!
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10))
with sns.axes_style("whitegrid"):

  cars = sns.boxplot(x=data['model_year'],y=data['weight'] )
  
  cars.set_title("Distribuição contínua aproximada do peso de carros por Modelo_Ano", fontsize=15,fontweight="bold")
  cars.set_xlabel("Modelo_Ano", fontsize=15);
  cars.set_ylabel("Peso", fontsize=15);

  # Insight do gráfico 3
  # Aqui vemos de um modo geral, que com o passar dos anos o peso dos carros diminuem em relação ao seu modelo!
```
