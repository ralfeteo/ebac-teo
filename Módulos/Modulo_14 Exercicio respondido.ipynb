## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
preco = data[["cut", "price"]].groupby("price")
preco.head()
```

```
diamonds = sns.relplot(data=data,x="price", y="carat", hue="cut",style="cut")

diamonds.ax.set_title("Preço do diamante por tipo de corte", fontsize=20, fontweight="bold");
diamonds.set_xlabels("Preço (USD)", fontsize=20);
diamonds.set_ylabels("Quilate", fontsize=20);
diamonds.legend.set_title("corte", prop={"size": 20, "weight": "bold"});
```

```
altura  = 15
largura = 19

diamonds.fig.set_size_inches(w=largura, h=altura)
diamonds.fig
```

```
diamonds = sns.relplot(data=data,x="price", y="carat", hue="cut",style="cut", col="cut")

diamonds.fig.suptitle("Preço do diamante por tipo de corte", fontsize=20, fontweight="bold"); 
diamonds.fig.subplots_adjust(top=.8);
diamonds.set_xlabels("Preço (USD)", fontsize=20);
diamonds.set_ylabels("Quilate", fontsize=20);
diamonds.legend.set_title("corte", prop={"size": 20, "weight": "bold"});
```

```
altura  = 15
largura = 25

diamonds.fig.set_size_inches(w=largura, h=altura)
diamonds.fig
# Aqui eu percebo que o diamante com o corte Ideal é o de maior quantidade, entre 1 e 2 quilates e também é o que mais se produz e vende mais!
```

```
diamonds.fig.savefig(fname="Preço do diamante por tipo de corte.png", bbox_inches="tight")
```

**Insight do gráfico 1**: ...

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10))

diamonds= sns.barplot(data=data,x="clarity", y="price", hue="cut",ci=None, palette="pastel")
diamonds.set(title="Preço do diamante por tipo de corte", xlabel='Clareza', ylabel='Preço');
diamonds.set_title("Preço do diamante por tipo de corte", fontsize=15)
diamonds.get_legend().set_title("Corte");
diamonds.set_xlabel("Clareza", fontsize=15); 
diamonds.set_ylabel("Preço", fontsize=15);

# Acho que esse gráfico de barras atende melhor a necessidade! Aqui mudei o eixo X para Clarity acho que fica melhor para analizar os preços!
# Fica melhor para ter uma visão dos preços pela Clareza!
```

```
diamonds.get_figure().savefig("gráfico_barra_diamonds.png")
```

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))

with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=data,x="clarity", y="price", hue="cut",ci=None, palette="pastel")
  grafico.set(title="Preço do diamante por tipo de corte", xlabel='Clareza', ylabel='Preço');
  grafico.set_title("Preço do diamante por tipo de corte", fontsize=15)
  grafico.get_legend().set_title("Corte");
  grafico.set_xlabel("Clareza", fontsize=15); 
  grafico.set_ylabel("Preço", fontsize=15);
  # Acho que esse gráfico de Linhas também atende a necessidade! Aqui mudei o eixo X para Clarity!
  #Esse tipo de gráfico também é bom para fazer uma análise de preços por clareza!
```

```
diamonds = data[data['cut'] == 'Fair']
diamonds.head()
```

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
flights = data[data['month'] == 'Dec']
flights
```

```
# gráfico do exercício 2
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,9))

grafico = sns.barplot(data=flights, x="year", y="passengers",hue="year",ci=None, palette="Paired")
grafico.set(title='Número de passageiros em dezembro por ano', xlabel='Ano', ylabel='Passageiros');
grafico.set_title('Número de passageiros em dezembro por ano',fontsize=15)
grafico.get_legend().set_title("Ano");
grafico.set_xlabel("Ano", fontsize=15); 
grafico.set_ylabel("Passageiros", fontsize=15);
plt.legend(loc='center left') #Achei esse método para por a legenda a esquerda!
# Acho que esse gráfico de barras atende a necessidade! Percebe-se que o número de passageiros vem crecendo a cada ano! 
```

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))

flights = data[data['month'] == 'Jan'] # Aqui basta mudar o mês para visualizar a barra do mês desejado!
grafico = sns.barplot(data=flights, x="year", y="passengers",hue="month",ci=None, palette="Paired")
grafico.set(title='Número de passageiros em Janeiro por ano', xlabel='Ano', ylabel='Passageiros');
grafico.set_title('Número de passageiros em Janeiro por ano',fontsize=15)
grafico.get_legend().set_title("Mês");
grafico.set_xlabel("Ano", fontsize=15); 
grafico.set_ylabel("Passageiros", fontsize=15);

# Analisando aqui o mês de janeiro para uma comparação!
```

```
#Salvando o gráfico
grafico.get_figure().savefig("gráfico_barra_mes_jan.png")
```

**Insight do gráfico 2**: ...

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
flights =  data.query("1949 <= year < 1960")
flights.head()
```

```
# gráfico do exercício 3
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(23,8))

with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=flights, x="month", y="passengers",hue="year",style="year", markers=True,dashes=False, palette="pastel")# Achei interessante essa configuração com marcadores!
  grafico.set(title='Passageiros por ano entre 1949 e 1960', xlabel='Mes', ylabel='Passageiros');
  grafico.set_title('Passageiros por ano entre 1949 e 1960',fontsize=15)
  grafico.get_legend().set_title("Ano");
  grafico.set_xlabel("Ano", fontsize=15); 
  grafico.set_ylabel("Passageiros", fontsize=15);

  
```

**Insight do gráfico 3**: ...

```
#Separei por ano específico para visualizar melhor a linha!
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))

with sns.axes_style('whitegrid'):
  flights =  data.query("1949 == year ")
  grafico = sns.lineplot(data=flights, x="month", y="passengers",hue="year", palette="pastel",markers=True,style="year")
  grafico.set(title='Passageiros por mês no ano de  1949', xlabel='Mes', ylabel='Passageiros');
  grafico.set_title('Passageiros por mês no ano de  1949',fontsize=15)
  grafico.get_legend().set_title("Ano");
  grafico.set_xlabel("Ano", fontsize=15); 
  grafico.set_ylabel("Passageiros", fontsize=15);
```
